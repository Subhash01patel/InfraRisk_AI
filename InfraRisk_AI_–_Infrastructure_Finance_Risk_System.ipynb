```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/supply_chain_data.csv")
```

```python
data.head()
```

```python
# feature engineering
data['project_cost'] = data['Manufacturing costs'] * 10
data['delay'] = data['Shipping times']
```

```python
# Macro features (manual)
data['interest_rate'] = 8
data['gdp_growth'] = 6
data['inflation'] = 5
```

```python
# ---------------- TARGET ----------------
data['infra_risk'] = (
    (data['delay'] > 7) |
    (data['inflation'] > 6)
).astype(int)
```

```python
# ---------------- FEATURES ----------------
features = ['project_cost', 'delay', 'interest_rate', 'gdp_growth', 'inflation']

X = data[features]
y = data['infra_risk']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model_ir = RandomForestClassifier(n_jobs=1)
model_ir.fit(X_train, y_train)
```

```python
y_pred = model_ir.predict(X_test)
print("InfraRisk Model Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pickle

with open("model_ir.pkl", "wb") as f:
    pickle.dump(model_ir, f)

print("model_ir.pkl saved ✅")
```
